Task 1 - Enter data manually and explore the variables

```python
import pandas as pd

# lists of country data
France = [16127955, 15805560, 17386612, 13960064, 4142050]
Germany = [15294292, 20447052, 23900423, 17988340, 5389106]
Italy = [10859356, 13275488, 18594885, 13299743, 4330074]
Spain = [9254305, 11015526, 14787185, 9182427, 2860952]
UK = [15746113, 17509234, 17554612, 13284207, 3437008]

# list of country and age groups
AgeGroups = ["0-19", "20-39", "40-59", "60-79", "80+"]
Countries = ['France', 'Germany', 'Italy', 'Spain', 'UK']

#create a dataframe from the data
dfCountryData = pd.DataFrame({'France':  France,
                           'Germany': Germany,
                           'Italy':   Italy,
                           'Spain':   Spain,
                           'UK':      UK, }, index = AgeGroups)
```

```python
# datframe meta data
display(dfCountryData.info())
display(dfCountryData.index)
display(dfCountryData.columns)
display(dfCountryData.size)
display(dfCountryData.shape)
display(dfCountryData.Italy)
display(dfCountryData.head(3))
display(dfCountryData.describe())
```

```python
import matplotlib.pyplot as plt

# for one country at a time display its age distribution
def plot_pie_chart(country):
  plt.figure(figsize=(5,5))
  plt.pie(dfCountryData[country], labels = dfCountryData.index)
  plt.title(f"Age distribution in {country}")

for country in Countries:
  plot_pie_chart(country)

```

```python
import matplotlib.pyplot as plt

# for one country at a time display its age distribution
def plot_bar_chart(country):
  plt.figure(figsize=(5,5))
  plt.bar(height =dfCountryData[country], x = dfCountryData.index)
  plt.title(f"Age distribution in {country}")

for country in Countries:
  plot_bar_chart(country)
```

```python
dfCountryData.plot.pie(subplots=True, figsize = (15,4), legend = False, autopct = '%1.1f')
plt.show()
dfCountryData.plot.bar(y = CountryData.columns, figsize = (15,4), legend = False, subplots = True, layout = (1,5))
plt.show()
```

```python
# (c) Clustered Bar Chart is the default when applying plot.bar to entire dataframe
dfCountryData.plot.bar(figsize = (12,4))
plt.show()
# to cluster by country take transpose of the dataframe (i.e. swap rows and columns)
display(dfCountryData.transpose())
dfCountryData.transpose().plot.bar(figsize = (12,4))
plt.show()
```

```python
# (d) Stacked Bar Chart
dfCountryData.plot.bar(figsize = (12,4), stacked = True)
plt.show()
# to cluster by country take transpose of the dataframe (i.e. swap rows and columns)
display(dfCountryData.transpose())
dfCountryData.transpose().plot.bar(figsize = (12,4), stacked = True)
plt.show()
```
